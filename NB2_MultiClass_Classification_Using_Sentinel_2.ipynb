```python
import numpy as np
import shutil
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from skimage.io import imread
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, confusion_matrix
```

```python
dataset = "EuroSAT_RGB"
```

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

train_data = []
val_data = []
test_data = []

for class_name in os.listdir(dataset):
    class_path = os.path.join(dataset, class_name)

    if not os.path.isdir(class_path):
        continue

    images = os.listdir(class_path)[:1024]
    train_images, test_images = train_test_split(images, test_size=0.2, random_state=42, stratify=None)
    train_images, val_images = train_test_split(train_images, test_size=0.125, random_state=42, stratify=None)

    for image in train_images:
        train_data.append({"image": os.path.join(class_path, image), "label": class_name})
    for image in test_images:
        test_data.append({"image": os.path.join(class_path, image), "label": class_name})
    for image in val_images:
        val_data.append({"image": os.path.join(class_path, image), "label": class_name})

train_df = pd.DataFrame(train_data)
val_df = pd.DataFrame(val_data)
test_df = pd.DataFrame(test_data)

train_df.to_csv(os.path.join(dataset, "train.csv"), index=False)
val_df.to_csv(os.path.join(dataset, "val.csv"), index=False)
test_df.to_csv(os.path.join(dataset, "test.csv"), index=False)
```

```python
train_df = pd.read_csv(os.path.join(dataset, "train.csv"))
val_df = pd.read_csv(os.path.join(dataset, "val.csv"))
test_df = pd.read_csv(os.path.join(dataset, "test.csv"))
unique_labels = np.unique(train_df["label"])
label_to_int = {label: i for i, label in enumerate(unique_labels)}
num_classes = len(unique_labels)
train_labels = np.array([label_to_int[label] for label in train_df["label"].values])
val_labels = np.array([label_to_int[label] for label in val_df["label"].values])
train_df.head()
```

```python
X_train = np.array([imread(image_path) for image_path in train_df["image"].values])
X_val = np.array([imread(image_path) for image_path in val_df["image"].values])
y_train = train_labels
y_val = val_labels
print(X_train.shape)
print(X_val.shape)
```

### Sample Images for Each Class

Let's visualize one sample image for each of the `EuroSAT_RGB` classes to get a better understanding of the dataset.

```python
import matplotlib.pyplot as plt
from skimage.io import imread

plt.figure(figsize=(15, 10))
for i, label in enumerate(unique_labels):
    # Find the first image path for the current label in the training set
    sample_image_path = train_df[train_df['label'] == label]['image'].iloc[0]

    # Load the image
    image = imread(sample_image_path)

    plt.subplot(2, 5, i + 1) # Arrange in a 2x5 grid for 10 classes
    plt.imshow(image)
    plt.title(label)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
clf = SVC(decision_function_shape='ovo', kernel = 'rbf')
clf.fit(X_train.reshape(X_train.shape[0],-1), y_train)
```

```python
y_train_pred = clf.predict(X_train.reshape(X_train.shape[0],-1))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate on training data
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
print("\nTraining Classification Report:\n", classification_report(y_train, y_train_pred, target_names=unique_labels))
print("\nTraining Confusion Matrix:\n", confusion_matrix(y_train, y_train_pred))

y_val_pred = clf.predict(X_val.reshape(X_val.shape[0],-1))

# Evaluate on validation data
print("\nValidation Accuracy:", accuracy_score(y_val, y_val_pred))
print("\nValidation Classification Report:\n", classification_report(y_val, y_val_pred, target_names=unique_labels))
print("\nValidation Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
```

### SVM with One-vs-Rest (OvR) Strategy

Now, let's train and evaluate an SVM using the One-vs-Rest (OvR) multiclass strategy. This approach trains a separate binary classifier for each class against all other classes.

```python
clf = SVC(decision_function_shape='ovr', kernel = 'rbf')
clf.fit(X_train.reshape(X_train.shape[0],-1), y_train)
```

```python
y_train_ovr_pred = clf.predict(X_train.reshape(X_train.shape[0],-1))

# Evaluate on training data
print("\n--- OVR SVM Training Performance ---")
print("Training Accuracy (OVR):", accuracy_score(y_train, y_train_ovr_pred))
print("\nTraining Classification Report (OVR):\n", classification_report(y_train, y_train_ovr_pred, target_names=unique_labels))
print("\nTraining Confusion Matrix (OVR):\n", confusion_matrix(y_train, y_train_ovr_pred))
```

```python
y_val_ovr_pred = clf.predict(X_val.reshape(X_val.shape[0],-1))

# Evaluate on validation data
print("\n--- OVR SVM Validation Performance ---")
print("Validation Accuracy (OVR):", accuracy_score(y_val, y_val_ovr_pred))
print("\nValidation Classification Report (OVR):\n", classification_report(y_val, y_val_ovr_pred, target_names=unique_labels))
print("\nValidation Confusion Matrix (OVR):\n", confusion_matrix(y_val, y_val_ovr_pred))
```

### Random Forest Model Training and Evaluation

Now, let's train and evaluate a Random Forest Classifier on the same dataset to compare its performance with the SVM.

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=25, random_state=42, n_jobs=-1) # n_jobs=-1 uses all available cores
rf_clf.fit(X_train.reshape(X_train.shape[0],-1), y_train)

print("Random Forest Classifier trained successfully.")
```

### Practical Comparison: SVM vs. Random Forest

Based on the evaluation results, we can observe the following practical differences:

*   **Training Time:** (Note: This is an observation based on general characteristics. For precise timings, you would need to use `time` module during training.) Random Forest models, especially with many estimators, can often take longer to train than a simple SVM, particularly on larger datasets, due to building multiple decision trees.

*   **Training Accuracy:** Random Forests often achieve higher training accuracy, sometimes even 100%, because they are powerful and can easily overfit the training data if not properly regularized.

*   **Validation/Generalization Performance:** The key difference lies in how well each model generalizes to unseen data (validation set).
    *   **SVM:** Depending on the kernel and parameters, SVMs can be robust, but might struggle with very complex, non-linear patterns without appropriate kernel tuning.
    *   **Random Forest:** Typically, Random Forests tend to generalize well due to the ensemble nature (bagging and feature randomness), which reduces variance and overfitting. However, they can still overfit if the individual trees are too deep or if there are too many estimators without enough diversity.

*   **Interpretability:** While not directly shown in metrics:
    *   **SVM:** Often considered a 'black box' for non-linear kernels, making it hard to understand individual predictions.
    *   **Random Forest:** While the overall ensemble is complex, individual decision trees are interpretable, and Random Forests can provide feature importances, offering insights into which features contribute most to the classification (though this is more relevant for tabular data than flattened images).

*   **Hyperparameter Tuning:**
    *   **SVM:** Crucial parameters like `C` (regularization) and `gamma` (kernel coefficient) significantly impact performance and require careful tuning.
    *   **Random Forest:** Key parameters include `n_estimators` (number of trees), `max_depth` (depth of trees), `min_samples_split`, and `max_features`. While less sensitive to scaling, tuning is still important for optimal performance and to prevent overfitting.

## SVM vs. Random Forest Models

Both Support Vector Machines (SVMs) and Random Forests are powerful and widely used supervised learning algorithms, but they operate on very different principles.

### Support Vector Machines (SVM)

**How it works:**
SVMs aim to find the optimal hyperplane that best separates data points of different classes in a high-dimensional space. The "optimal" hyperplane is the one with the largest margin between the closest data points of different classes (called support vectors).

*   **Linear SVM:** Works well when data can be separated by a straight line or plane.
*   **Non-linear SVM (Kernel Trick):** Uses kernel functions (like RBF, polynomial) to transform the input data into a higher-dimensional space where it might become linearly separable. This allows SVMs to handle complex, non-linear relationships.
*   **Multiclass Strategies:** For problems with more than two classes, SVMs typically employ strategies like:
    *   **One-vs-One (OvO):** Trains a binary classifier for every pair of classes. The class that gets the most votes is chosen. (e.g., `decision_function_shape='ovo'` in scikit-learn).
    *   **One-vs-Rest (OvR) / One-vs-All (OvA):** Trains a binary classifier for each class against all other classes. The class with the highest score is selected. (e.g., `decision_function_shape='ovr'` in scikit-learn).

**Strengths:**
*   **Effective in high-dimensional spaces:** Especially when the number of dimensions is greater than the number of samples.
*   **Memory efficient:** Uses a subset of training points (support vectors) in the decision function.
*   **Versatile:** Different kernel functions can be specified for the decision function.

**Weaknesses:**
*   **Slow for large datasets:** Training time can be long.
*   **Sensitive to feature scaling:** Features should be normalized or standardized.
*   **Difficult to interpret:** The models are often considered "black boxes" and it can be hard to understand why a particular classification was made.
*   **Choice of kernel and parameters:** Performance is highly dependent on the choice of kernel function and regularization parameters (e.g., C, gamma).

### Random Forest

**How it works:**
Random Forest is an ensemble learning method that builds multiple decision trees during training and outputs the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees. It introduces randomness in two main ways:

1.  **Bagging (Bootstrap Aggregating):** Each tree is trained on a different random subset of the training data (with replacement).
2.  **Feature Randomness:** For each split in a decision tree, only a random subset of features is considered.

This combination reduces overfitting and improves the model's generalization ability.

**Strengths:**
*   **High accuracy:** Often achieves very good performance.
*   **Robust to overfitting:** Due to the ensemble nature and randomness.
*   **Handles large datasets with high dimensionality:** Can manage a large number of features and samples.
*   **Handles missing values:** Can handle missing values implicitly.
*   **Less sensitive to feature scaling:** Doesn't require features to be scaled.
*   **Provides feature importance:** Can estimate the importance of each feature in the decision-making process.

**Weaknesses:**
*   **Less interpretable:** While individual trees are interpretable, the ensemble itself can be difficult to visualize and understand.
*   **Computationally intensive:** Building many trees can be slow, especially with very large datasets.
*   **Memory intensive:** Requires more memory to store all the trees.

### When to Choose Which:

*   **Choose SVM when:**
    *   You have a relatively smaller dataset with a clear margin of separation.
    *   You need a powerful model for classification and can manage feature scaling and parameter tuning.
    *   You are working in high-dimensional spaces where the number of features is greater than the number of samples.

*   **Choose Random Forest when:**
    *   You have a large dataset and want a robust, high-accuracy model.
    *   You need an algorithm that is less sensitive to feature scaling and missing data.
    *   You want to understand feature importance.
    *   You prefer a model that is generally less prone to overfitting and easier to tune.
